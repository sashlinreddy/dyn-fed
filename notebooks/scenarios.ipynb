```python
import pandas as pd
```

```python
enc_vars = [
        "n_workers", "scenario", "remap", "quantize",
        "comm_period", "send_gradients", "mu_g", "n_most_rep"
    ]
```

```python
n_workers = [99]
scenario = [1]
remap = [1]
quantize = [0, 1]
comm_period = [1, 10]
send_g = [0, 1]
mu_g = [0.0, 1.0]
n_most_rep = [100]
```

```python
perms = [[i,j,k, l, m, n, o, p] for i in n_workers for j in scenario for k in remap for l in quantize for m in comm_period for n in send_g for o in mu_g for p in n_most_rep]
```

```python
df = pd.DataFrame(perms, columns=enc_vars)
```

```python
df.loc[15, "slurm_id"] = 9192
```

```python
df.loc[15, "wall_time_seconds"] = 21.457917
```

```python
df.loc[11, "accuracy"] = 83.39
```

```python
df
```

```python
df.sort_values(by=["accuracy", "wall_time_seconds"], ascending=[False, True])
```

* n_workers: No. of workers
* scenario: Scenario type
* remap: Type of remap strategy when workers die (0 - remap all, 1 - only remap most representative points of dead worker(s))
* quantize: Whether or not to quantize parameters
* comm_period: Communication period (Every ith iteration)
* send_gradients: Whether or not to send gradients or local parameters
* mu_g: Parameter that penalizes local parameters so that the search space does not explore too far from the global parameters
* n_most_rep: No. of most representative data points saved for when the worker dies
* slurm_id: slurm job id
* accuracy: Test accuracy
* wall_time_seconds: Time it took for training

```python
df.to_csv("../data/results.csv", index=False)
```
